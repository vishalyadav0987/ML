#### Map Function in Python

```python
def even_odd(num):
    if num%2==0:
        return ("The given number is {} even.".format(num))
    else:
        return ("The given number is {} odd.".format(num))
```

```python
lst = [1,2,3,4,5,6,7,8,9]
```

```python
# One way to check odd or even from list using loop
for num in lst:
    print(even_odd(num=num))
```

```python
# Another way is to iterate is using 'Map'
list(map(even_odd,lst)) # function,list
```
