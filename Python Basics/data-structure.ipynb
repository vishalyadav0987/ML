#### *Boolean Variables*
- Boolean values are the two constant objects True anf False.
- They are used to represt truth values.
- In numeric contexts for example, when used as the argument to an arithmetic operator they behave like the integers 0 and 1 respectively.
- The built-in function bool() can be used cast any value to a boolean if the value can be interpreted as a truth value.


```python
print(True,False)
```

```python
type(True)
```

```python
bool()
```

```python
bool(True)
```

```python
my_var = 'Vishal'
```

```python
print(my_var.isalnum()) # check if the all char are numbers
print(my_var.isalpha()) # check if all char in the string are alphabets
print(my_var.isdigit()) # test if string contains digit
print(my_var.istitle()) # test if string contain title word (Vishal Yadav)
print(my_var.isupper()) # test A string is uppercase if all cased characters in the string
print(my_var.isspace()) # test if string contain spaces
print(my_var.endswith('v')) # case matter
print(my_var.startswith('v')) # case matter
```

#### SETS
A Set is an unorderd collection data type that is itrable, mutable, and has no duplicate elements. Pyhton's set class represents the mathematical notion of a set. This is based on data structure known as a hash table.

```python
# defining of empty set
set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var={1,2,3,4,5,'flutter'}
set_var
```

```python
set_var[0]
set_var['flutter'] #'set' object is not subscriptable
```

```python
# Ibuilt function in set
set_var.add('reactjs')
set_var
```

```python
set1 = {'mongo','express','react','node','css'}
set2 = {'react','flutter','golang','ML','css'}
```

```python
# differnce
set2.difference(set1)
```

```python
set2 # not change anthing when we use 'difference' in set2
```

```python
# differnce_update
set2.difference_update(set1)
```

#### Dictionaries
A Dictionaries is a collection which is unorderd, changeable, and indexed, In python dictionary are written with curly brackets, and they have keys and values.

```python
dic={}
type(dic)
```

```python
dic = dict()
type(dic)
```

```python
my_dict = {
    'name':'vishal',
    'age':22,
    'class':12
}
my_dict
```

```python
type(my_dict)
```

```python
# Access the vlaue from dictionary
my_dict['name']
```

```python
my_dict['name'] = 'vishal yadav'
# it will over write the previos exsisting key
```

```python
my_dict[1]='new_value'
# if key not present in dict it will append it
```

```python
my_dict
```

```python
# loop through dictionary key
for keys in my_dict:
    print(keys)
```

```python
# loop through dictionary values
for values in my_dict.values():
    print(values)
```

```python
# loop through dictionary another way to get value
for keys in my_dict:
    print(my_dict[keys])
```

```python
# we can also iterate both in tuple
for x in my_dict.items():
    print(type(x))
    print(x)
```

```python
# nested dictionary
skill1 = {'language':'reactjs'}
skill2 = {'language':'flutter'}
skill3 = {'language':'rnodejs'}

skill_type = {
    'mern':skill1,
    'new':skill2,
    'old_one':skill3
}
```

```python
skill_type
```

```python
print(skill_type['old_one']['language'])
```
