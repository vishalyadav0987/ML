#### List Comprehension
List Comprehension provide a concise way to crealte list. It contains of brakets containing and expression followed by a for clause then zero or more for or if caluses. The expression can be anything, meaning you can put in all kinds of objects in lists

**List Comprehension**
List comprehension Python me **list banane ka short & clean way** hota hai.
Isme **square brackets** ke andar pehle **expression** hota hai, phir **for loop**, aur optional **if condition** hoti hai.
Expression kuch bhi ho sakta hai, matlab tum **kisi bhi type ke objects** list me daal sakte ho.

**Simple example:**

```python
squares = [x*x for x in range(5)]
```

Yahan har `x` ka square nikal ke list ban rahi hai.


````python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
````

````python
lst_square([1,2,3,4,5,6]) # these is longer way
````

````python
# list comprehension is sorter way to create list with, condioton and loop
list_square = [i*i for i in range(5)]
````

````python
print(list_square)
````

````python
list_square = [i*i for i in range(5) if i%2==0]
print(list_square)
````
