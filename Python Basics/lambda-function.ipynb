#### Lanmbda Functions
- Also Called anonymous (no named) funtion.


```python
def sum(a,b):
    print("The sum of {} + {} = ".format(a,b),a+b)
```

```python
sum(24,36)
```

```python
# The Sort Way to do using Lambda function
sum = lambda a,b:a+b
```

```python
print(sum(24,36))
```

```python
even_odd = lambda a:a%2==0
```

```python
even_odd(5)
```

```python
even_odd(4)
```

```python
def additions(x,y,z):
    return x+y+z
```

```python
print(additions(24,56,76))
```

```python
# Using Lambda
additions = lambda x,y,z:x+y+z
```

```python
print(additions(24,36,76))
```
